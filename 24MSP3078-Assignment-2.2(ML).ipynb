```python
import statsmodels.api as sm
X=data.drop(['Heating_Load'] , axis=1)
```

```python
# Drop features with perfect multicollinearity
X = X.drop(columns=['Surface_Area', 'Wall_Area', 'Roof_Area'])

# Recompute VIF
X_const = sm.add_constant(X)
vif_feature = pd.DataFrame()
vif_feature["Features"] = X_const.columns
vif_feature["VIF"] = [variance_inflation_factor(X_const.values, i) for i in range(X_const.shape[1])]

# Sort and display
vif_feature.sort_values(by="VIF", inplace=True)
print(vif_feature)
```

```python
from sklearn.model_selection import train_test_split
Y=data['Heating_Load']
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2 , random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
f_score,p_value = f_regression(X_train , Y_train)
#wrapper method:
rfe=RFE(LinearRegression(),n_features_to_select=7)
rfe.fit(X_train,Y_train)
feature_selected=pd.DataFrame({
    'Features' : X_train.columns,
    'F_score' : f_score,
    'P-value': p_value,
    'RFE Ranking': rfe.ranking_
})
feature_selected.sort_values(by='RFE Ranking')
```

```python
f_score,p_value = f_regression(X_train , Y_train)
#wrapper method:
rfe=RFE(LinearRegression(),n_features_to_select=7)
rfe.fit(X_train,Y_train)
feature_selected=pd.DataFrame({
    'Features' : X_train.columns,
    'F_score' : f_score,
    'P-value': p_value,
    'RFE Ranking': rfe.ranking_
})
feature_selected.sort_values(by='F_score')
```

```python
print(X_train.columns)
print(rfe.support_)
```

```python
#embebbed methods :
from sklearn.linear_model import  Ridge , Lasso
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
# Lasso model building:
lasso = Lasso(alpha=0.5)  # alpha is the regularization parameter
lasso.fit(X_train, Y_train)
y_pred_lasso= lasso.predict(X_test)

sns.scatterplot(x=y_pred_lasso, y=Y_test , color = 'blue', label = 'Lasso Regression')

#evalution:
# Evaluate the model
mse_lasso= mean_squared_error(Y_test, y_pred_lasso)
print("Mean Squared Error ( Lasso) :", mse_lasso)
r2=r2_score(Y_test , y_pred_lasso)
print('r2 is : ', r2)
```

```python
#ridge:
# Ridge model building : as multicollinerity exist in our model:
ridge_model=Ridge(alpha=1.0)
ridge_model.fit(X_train , Y_train)
y_pred_ridge=ridge_model.predict(X_test)
sns.scatterplot(x=y_pred_ridge, y=Y_test , color = 'red', label = 'Ridge Regression')

#Evalution:
mse_ridge= mean_squared_error(Y_test, y_pred_ridge)
print("Mean Squared Error (Ridge) :", mse_ridge)
r2=r2_score(Y_test , y_pred_ridge)
print('r2 is : ', r2)
```

```python
#linear regression using all features:
# model building : linear regression:

linear_model=LinearRegression()
linear_model.fit(X_train , Y_train)
y_pred = linear_model.predict(X_test)
```

```python
# perfomance evaluation:

r2=r2_score(Y_test , y_pred)
print('r2 is : ', r2)

mse=mean_squared_error(Y_test , y_pred)
print('MSE is : ',mse)

mae = mean_absolute_error(Y_test , y_pred)
print('MAE is : ', mae)

rmse = np.sqrt(mse)
print('RMSE is : ',rmse)
```

```python
# Best features identified by each method
filter_features = ["Glazing_Area","Wall_Area" , "Overall_Height", "Roof_Area", "Surface_Area","Relative_Compactness","Cooling_Load"]  # Based on f_regression
rfe_features = [ "Roof_Area", "Surface_Area", "Relative_Compactness", "Wall_Area","Glazing_Area" , "Overall_Height" , "Cooling_Load"]  # Based on RFE


# Train and evaluate models for Heating Load
linear_model_filter_heating = LinearRegression()
linear_model_filter_heating.fit(X_train[filter_features], Y_train)
y_pred_filter_heating = linear_model_filter_heating.predict(X_test[filter_features])

linear_model_rfe_heating = LinearRegression()
linear_model_rfe_heating.fit(X_train[rfe_features], Y_train)
y_pred_rfe_heating = linear_model_rfe_heating.predict(X_test[rfe_features])


# Evaluate and compare models
def evaluate_results(y_true, y_pred, target_name):
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    return {"Target": target_name, "MSE": mse, "MAE": mae, "R2": r2}

results = pd.DataFrame([
    evaluate_results(Y_test, y_pred_filter_heating, "Heating_Load (Filter)"),
    evaluate_results(Y_test, y_pred_rfe_heating, "Heating_Load (RFE)"),
])

print(results)
```

---> 1. Features Selected by Each Method :

--->Filter method:
..Features selected: Glazing_Area, Wall_Area, Overall_Height, Roof_Area, Surface_Area, Relative_Compactness, Cooling_Load.
..Reason for selection: The filter method identifies features based on their statistical correlation with the target variable (Heating_Load).
                        It evaluates each feature independently and selects the ones most strongly correlated using techniques like f_regression.

--->RFE method:
..Features selected: Roof_Area, Surface_Area, Relative_Compactness, Wall_Area, Glazing_Area, Overall_Height, Cooling_Load.
..Reason for selection: Recursive Feature Elimination (RFE) iteratively trains a model, ranks the features, and removes the least significant one until the optimal subset is reached.
                       This method ensures the selected features maximize the model's predictive accuracy.


--->2. Impact of Feature Selection on Model Performance:

--->Performance Metrics:

..Filter Method:

    Mean Squared Error (MSE): 3.885
    Mean Absolute Error (MAE): 1.408
    R Score: 0.9627


..RFE Method:

    Mean Squared Error (MSE): 3.826
    Mean Absolute Error (MAE): 1.384
    R Score: 0.9633


--->Analysis:
Both methods resulted in high R scores (> 0.96), indicating the models have strong predictive capabilities.
The RFE model demonstrated slightly better performance, with lower MSE and MAE and a marginally higher R score.
This shows that the features selected by RFE better capture the relationship between the predictors and Heating_Load.


--->3. Which Feature Selection Technique Worked Best and Why?
The RFE method worked better for this dataset, as evidenced by its improved performance metrics.
The reasons can be :

It evaluates features in the context of their overall contribution to the model, considering interactions and dependencies between variables.
Unlike the Filter method, which relies solely on correlation, RFE identifies features that collectively enhance the models predictive accuracy.
The slight improvement in MSE, MAE, and R values reflects the superior quality of features selected by RFE.

Note: Even the filter method is provding almost similar accuracy .

--->Summary:

..Filter Method is simple and computationally efficient, making it suitable for quick analysis, but it doesnt account for feature interactions.
..RFE Method is more thorough, as it incorporates feature importance during model training, ensuring a more refined feature subset.
..Here RFE is the preferred feature selection technique, as it delivers marginally better performance and captures feature relationships effectively.
